```python
import tiktoken
from IPython.display import display, HTML
import random
```

```python
text = "unhappy"
print("Input text:")
print(text)
```

```python
encoding = tiktoken.encoding_for_model("gpt-4o-mini")
```

```python
tokens = encoding.encode(text)
tokens
```

```python
decoded_tokens = [encoding.decode([token]) for token in tokens]
decoded_tokens
```

```python
char_count = len(text)
token_count = len(tokens)
ratio = char_count / token_count

print("Character count:", char_count)
print("Token count:", token_count)
print("Character-to-token ratio:", round(ratio, 2))
```

```python
colors = [
    "#FFCDD2", "#F8BBD0", "#E1BEE7", "#D1C4E9",
    "#C5CAE9", "#BBDEFB", "#B2EBF2", "#C8E6C9",
    "#DCEDC8", "#FFF9C4", "#FFE0B2"
]

html_output = ""

for token in decoded_tokens:
    color = random.choice(colors)
    html_output += f"<span style='background-color:{color}; padding:5px; margin:2px; border-radius:5px;'>{token}</span>"

display(HTML(html_output))
```
